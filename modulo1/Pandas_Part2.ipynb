## Filtrando Dados

Podemos filtrar as linhas do data frame baseando-se em condições.

Trazer todas as linhas cujos pokémons são do Type 1 = Grass

````python
#Primeiro conheça os tipos existentes nessa coluna
df_pokemons['Type 1']
````

````python
#filtrar pelo Type 1 == 
df_pokemons.loc[df_pokemons['Type 1']=='Grass']
````

### Exercício - FAzer sozinho!

### Trazer todas as linhas cujos pokémons são do Type 1 = Grass E Type 2 = 'Poison'

### AND

````python
# condições múltiplas
# Pokemon Type 1 = Grass E Type 2 = Poison
df_pokemons.loc[(df_pokemons['Type 1']=='Grass') & (df_pokemons['Type 2']=='Poison')]
````

### OR

````python
#Ou
# Pokemon Type 1 = Grass OU Type 2 = Poison
df_pokemons.loc[(df_pokemons['Type 1']=='Grass') | (df_pokemons['Type 2']=='Poison')]
````

## Mudança Condicional

Baseando-se nas condições que desejarmos, podemos substituir os valores de células.

## Exercicio - fazer junto

 Todo Type 1 de qualquer pokémon cujo HP seja maior que 60 será "LENDÁRIO"

````python
# Alterando um único valor
df_pokemons.loc[df_pokemons['HP']>60,'Type 1'] = 'LENDÁRIO'
````
